```python
import sys
sys.path.append('../src')

from data_loader import load_data
from data_cleaning import clean_data

# Carregar os dados
df = load_data('../data/raw/Customer_support_data.csv')

# Limpar os dados
df_clean = clean_data(df)

# Exibir as primeiras linhas
df_clean.head()
```

```python
from data_analysis import basic_statistics, plot_satisfaction_distribution
```

```python
basic_statistics(df_clean)
plot_satisfaction_distribution(df_clean)
```

```python
# Verificar valores ausentes
df_clean.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.countplot(x="csat_score", data=df_clean, palette="viridis")
plt.title("Distribuição da Satisfação dos Clientes")
plt.xlabel("Nota de Satisfação")
plt.ylabel("Quantidade")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="csat_score", y="connected_handling_time", data=df_clean, palette="coolwarm")
plt.title("Tempo de Atendimento vs. Nota de Satisfação")
plt.xlabel("Nota de Satisfação")
plt.ylabel("Tempo de Atendimento (segundos)")
plt.show()
```

```python
df_clean["csat_score"].value_counts()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="product_category", y="csat_score", data=df_clean, palette="magma", ci=None)
plt.xticks(rotation=90)
plt.title("Satisfação Média por Categoria de Produto")
plt.xlabel("Categoria do Produto")
plt.ylabel("Média da Nota de Satisfação")
plt.show()
```

```python
df_clean["product_category"].value_counts()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="channel_name", y="csat_score", data=df_clean, palette="Set2")
plt.xticks(rotation=45)
plt.title("Satisfação por Canal de Atendimento")
plt.xlabel("Canal de Atendimento")
plt.ylabel("Nota de Satisfação")
plt.show()
```

```python
df_clean["channel_name"].value_counts()
```

```python
print("📊 Relatório de Análise de Satisfação do Cliente 📊\n")

# 1️⃣ Distribuição das Notas de Satisfação
print("🔹 Distribuição das notas de satisfação:")
print(df_clean["csat_score"].value_counts(), "\n")

# 2️⃣ Tempo de Atendimento vs Satisfação
print("🔹 Tempo médio de atendimento por nota de satisfação:")
print(df_clean.groupby("csat_score")["connected_handling_time"].mean(), "\n")

# 3️⃣ Satisfação por Categoria de Produto
print("🔹 Média de satisfação por categoria de produto:")
print(df_clean.groupby("product_category")["csat_score"].mean(), "\n")

# 4️⃣ Satisfação por Canal de Atendimento
print("🔹 Média de satisfação por canal de atendimento:")
print(df_clean.groupby("channel_name")["csat_score"].mean(), "\n")

# 5️⃣ Quantidade de atendimentos por canal
print("🔹 Número de atendimentos por canal:")
print(df_clean["channel_name"].value_counts(), "\n")
```
