```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.countplot(x="csat_score", data=df_clean, palette="viridis")
plt.title("DistribuiÃ§Ã£o da SatisfaÃ§Ã£o dos Clientes")
plt.xlabel("Nota de SatisfaÃ§Ã£o")
plt.ylabel("Quantidade")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="csat_score", y="connected_handling_time", data=df_clean, palette="coolwarm")
plt.title("Tempo de Atendimento vs. Nota de SatisfaÃ§Ã£o")
plt.xlabel("Nota de SatisfaÃ§Ã£o")
plt.ylabel("Tempo de Atendimento (segundos)")
plt.show()
```

```python
df_clean["csat_score"].value_counts()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="product_category", y="csat_score", data=df_clean, palette="magma", ci=None)
plt.xticks(rotation=90)
plt.title("SatisfaÃ§Ã£o MÃ©dia por Categoria de Produto")
plt.xlabel("Categoria do Produto")
plt.ylabel("MÃ©dia da Nota de SatisfaÃ§Ã£o")
plt.show()
```

```python
df_clean["product_category"].value_counts()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="channel_name", y="csat_score", data=df_clean, palette="Set2")
plt.xticks(rotation=45)
plt.title("SatisfaÃ§Ã£o por Canal de Atendimento")
plt.xlabel("Canal de Atendimento")
plt.ylabel("Nota de SatisfaÃ§Ã£o")
plt.show()
```

```python
df_clean["channel_name"].value_counts()
```

```python
print("ðŸ“Š RelatÃ³rio de AnÃ¡lise de SatisfaÃ§Ã£o do Cliente ðŸ“Š\n")

# 1ï¸âƒ£ DistribuiÃ§Ã£o das Notas de SatisfaÃ§Ã£o
print("ðŸ”¹ DistribuiÃ§Ã£o das notas de satisfaÃ§Ã£o:")
print(df_clean["csat_score"].value_counts(), "\n")

# 2ï¸âƒ£ Tempo de Atendimento vs SatisfaÃ§Ã£o
print("ðŸ”¹ Tempo mÃ©dio de atendimento por nota de satisfaÃ§Ã£o:")
print(df_clean.groupby("csat_score")["connected_handling_time"].mean(), "\n")

# 3ï¸âƒ£ SatisfaÃ§Ã£o por Categoria de Produto
print("ðŸ”¹ MÃ©dia de satisfaÃ§Ã£o por categoria de produto:")
print(df_clean.groupby("product_category")["csat_score"].mean(), "\n")

# 4ï¸âƒ£ SatisfaÃ§Ã£o por Canal de Atendimento
print("ðŸ”¹ MÃ©dia de satisfaÃ§Ã£o por canal de atendimento:")
print(df_clean.groupby("channel_name")["csat_score"].mean(), "\n")

# 5ï¸âƒ£ Quantidade de atendimentos por canal
print("ðŸ”¹ NÃºmero de atendimentos por canal:")
print(df_clean["channel_name"].value_counts(), "\n")
```
